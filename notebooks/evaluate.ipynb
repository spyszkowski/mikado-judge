# Mikado Judge — Model Evaluation

Evaluates a trained YOLO-OBB model on the validation set and computes
per-class precision, recall, and mAP metrics.

```python
from google.colab import drive
drive.mount('/content/drive')

DRIVE_ROOT = '/content/drive/MyDrive/mikado-judge'
WEIGHTS_PATH = f'{DRIVE_ROOT}/weights/mikado_v1_best.pt'  # Update
DATASET_DIR = f'{DRIVE_ROOT}/datasets/mikado'
IMGSZ = 640
```

```python
!pip install ultralytics --quiet
```

```python
from ultralytics import YOLO
import yaml

# Write dataset yaml
dataset_yaml = {
    'path': DATASET_DIR,
    'train': 'images/train',
    'val': 'images/val',
    'nc': 5,
    'names': {0: 'mikado', 1: 'blue', 2: 'red', 3: 'yellow', 4: 'green'},
}
yaml_path = f'{DATASET_DIR}/dataset.yaml'
with open(yaml_path, 'w') as f:
    yaml.dump(dataset_yaml, f)

model = YOLO(WEIGHTS_PATH)
metrics = model.val(data=yaml_path, task='obb', imgsz=IMGSZ)

print(f'\n=== Validation Results ===')
print(f'mAP50:    {metrics.box.map50:.4f}')
print(f'mAP50-95: {metrics.box.map:.4f}')
print(f'Precision: {metrics.box.mp:.4f}')
print(f'Recall:    {metrics.box.mr:.4f}')
```

```python
# Per-class metrics
names = ['mikado', 'blue', 'red', 'yellow', 'green']
print('\nPer-class AP50:')
for i, (name, ap) in enumerate(zip(names, metrics.box.ap50)):
    print(f'  {name:10s}: {ap:.4f}')
```
